```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
#creating the data

np.random.seed(42)
n=1000

departments=['HR','Engineering','Sales','Marketing','Finance', 'Operations']

data = {
    'Age': np.random.randint(22, 60, n),
    'Department': np.random.choice(departments, n),
    'Experience_Years': np.random.randint(0, 35, n),
    'Training_Hours': np.random.randint(5, 120, n),
    'Performance_Score': np.round(np.random.normal(3.5, 0.7, n), 2),
    'Salary': np.random.randint(300000, 1800000, n),
    'Work_Life_Balance': np.random.randint(1, 6, n)  # 1 (poor) to 5 (excellent)
}

df = pd.DataFrame(data)

# Ensure realistic bounds
df['Performance_Score'] = df['Performance_Score'].clip(1, 5)

df.head()
```

```python
plt.figure()
sns.countplot(x='Department', data=df)
plt.title("Employee Distribution per Department")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure()
sns.barplot(x='Department', y='Salary', data=df, estimator=np.mean)
plt.title("Average Salary per Department")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure()
sns.barplot(x='Department', y='Training_Hours', data=df, estimator=np.mean)
plt.title("Average Training Hours by Department")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Training_Hours', y='Performance_Score', hue='Department', data=df)
plt.title("Training Hours vs Performance Score")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Department', y='Work_Life_Balance', data=df)
plt.title("Department-wise Work-Life Balance")
plt.xticks(rotation=30)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("indian_food.csv")

df.head()
```

```python
plt.figure()
sns.countplot(x='diet', data=df)
plt.title("Count of Vegetarian and Non-Vegetarian Dishes")
plt.xlabel("Diet Type")
plt.ylabel("Count")
plt.show()
```

```python
unique_dishes = df['name'].nunique()
print("Number of unique dishes:", unique_dishes)
```

```python
flavor_counts = df['flavor_profile'].value_counts()

plt.figure()
sns.barplot(x=flavor_counts.index, y=flavor_counts.values)
plt.title("Count of Dishes by Flavor Profile")
plt.xlabel("Flavor Profile")
plt.ylabel("Number of Dishes")
plt.xticks(rotation=30)
plt.show()
```
