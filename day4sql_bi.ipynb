```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("Electric_Vehicle_Population_Data.csv")
```

```python
df.info()
```

```python
df.drop_duplicates()
```

```python
total_vehicles = len(df)
avg_range = df["Electric Range"].mean()
print("Total Vehicles:", total_vehicles)
print("Avg Electric Range:", round(avg_range,2))
```

```python
bev_count = df["Electric Vehicle Type"].str.contains("BEV").sum()
phev_count = df["Electric Vehicle Type"].str.contains("PHEV").sum()
```

```python
bev_pct = bev_count / total_vehicles * 100
phev_pct = phev_count / total_vehicles * 100


print("BEV %:", round(bev_pct,2))
print("PHEV %:", round(phev_pct,2))
```

```python
year_df = (
    df[df["Model Year"] >= 2010]
    .groupby("Model Year")
    .size()
    .reset_index(name="Total Vehicles")
)

fig_year = px.area(
    year_df,
    x="Model Year",
    y="Total Vehicles",
    title="Total Vehicles by Model Year (From 2010 Onwards)"
)

fig_year.show()
```

```python
state_df = (
    df.groupby("State")
    .size()
    .reset_index(name="Total Vehicles")
)

fig_state = px.choropleth(
    state_df,
    locations="State",
    locationmode="USA-states",
    color="Total Vehicles",
    scope="usa",
    title="Total Vehicles by State",
    color_continuous_scale="Blues"
)

fig_state.show()
```

```python
make_df = (
    df.groupby("Make")
    .size()
    .reset_index(name="Total Vehicles")
    .sort_values(by="Total Vehicles", ascending=False)
    .head(10)
)

fig_make = px.bar(
    make_df,
    x="Make",
    y="Total Vehicles",
    title="Top 10 Total Vehicles by Make",
    text="Total Vehicles"
)

fig_make.show()
```

```python
cafv_df = (
    df.groupby("Clean Alternative Fuel Vehicle (CAFV) Eligibility")
    .size()
    .reset_index(name="Total Vehicles")
)

fig_cafv = px.pie(
    cafv_df,
    names="Clean Alternative Fuel Vehicle (CAFV) Eligibility",
    values="Total Vehicles",
    hole=0.4,
    title="Total Vehicles by CAFV Eligibility"
)

fig_cafv.show()
```

```python
model_df = (
    df.groupby("Model")
    .size()
    .reset_index(name="Total Vehicles")
    .sort_values(by="Total Vehicles", ascending=False)
    .head(10)
)

fig_model = px.treemap(
    model_df,
    path=["Model"],
    values="Total Vehicles",
    title="Top 10 Electric Vehicle Models"
)

fig_model.show()
```
